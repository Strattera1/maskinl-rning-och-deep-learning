**Dela upp datan i träning och testset**

```python
X = df.drop(columns=['quality','Id'])
Y = df['quality']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

X_train.reshape(-1,1)

Y_train.reshape(1,-1)

```python
model = DecisionTreeClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
model.score(X_train, Y_train)
```

**Vad får model för test score?**

```python
model.score(X_test, Y_test)
```

**Hur djupt är trädet?**

```python
model.get_depth()
```

**Hur många löv har trädet?**

```python
model.get_n_leaves()
```

```python
X = df.drop(columns=['quality','Id'])
Y = df['quality']
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
model_test = DecisionTreeClassifier()
```

```python
model_test.fit(X_train, Y_train)
```

```python
model_test.score(X_train,Y_train)
```

```python
model_test.score(X_test,Y_test)
```

```python
model_test.get_depth()
```

```python
model_test.get_n_leaves()
```

**Hur skiljer de sig åt och varför?**

*Svar*:
Från de testerna jag har gjort, har jag märkt att ovasett hur mycket man ändrar på test_size och random_state, blir tränings datan alltid 1.0, medan testdatan blir påverkat av de och ändra sig. Det gäller även djupet på trädet och hur många löv det blir, om man inte har satt en parameter för hur djup det ska vara och hur många löv det ska vara. 

**Hur kan man göra något åt fenomenet?**

*svar:*
Om träningsdatan ska också ändra sig ifrån 1.0, då behöver man använda sig av max depth.
