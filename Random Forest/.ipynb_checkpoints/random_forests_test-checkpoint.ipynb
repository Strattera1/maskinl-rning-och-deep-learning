```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import random
```

```python
df = pd.read_csv ('WineQT.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=['quality', 'Id'])
Y = df['quality']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
def train_forest(inputs, target, trees= 10):
    forest = []
    attributes = inputs.columns
    for i in range (trees):
        tree = DecisionTreeClassifier()
        train_columns = random.sample(
        list(attributes),
            random.randint(1, len(attributes))
        )
        tree.fit(inputs[train_columns], target)
        forest.append({"model": tree, "columns": train_columns})
        
    return forest
```

```python
rf = train_forest (X_train, Y_train)
```
