```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import xgboost as xgb
```

```python
df  = pd.read_csv ("students_adaptability_level_online_education.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

**Sklearn API**

```python
def encode_columns(df):
    columns = df.columns
    for c in columns:
        le = LabelEncoder()
        le.fit(df[c])
        df[c] = le.transform(df[c])
    return df
```

```python
encoded_df = encode_columns(df)
```

```python
encoded_df.dtypes
```

```python
encoded_df.head()
```

```python
X = encoded_df.drop(columns="Adaptivity Level")
y = encoded_df["Adaptivity Level"]
```

```python
X_train,X_test,y_train,y_test = train_test_split (X, y, test_size=0.3, random_state=42)
```

```python
clf = xgb.XGBClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
```

```python
accuracy_score(y_test,predictions)
```

**Learning API**

```python
df = pd.read_csv("students_adaptability_level_online_education.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
def encode_columns(df):
    columns = df.columns
    for c in columns:
        le = LabelEncoder()
        le.fit(df[c])
        df[c]= le.transform(df[c])
    return df
```

```python
encoded_df = encode_columns(df)
```

```python
df.head()
```

```python
X = encoded_df.drop(columns="Adaptivity Level")
y = encoded_df["Adaptivity Level"]
```

```python
X_train,X_test,y_train,y_test = train_test_split (X, y, test_size=0.3, random_state=42)
```

```python
m_X_Train = xgb.DMatrix(data = X_train, label = y_train)
```

```python
clf = xgb.train(
params={
    "objective": "multi:softmax",
    "num_class": 3,
    "max_depth": 10
},
dtrain = m_X_Train
)
```

```python
clf
```

```python
m_X_test = xgb.DMatrix (data = X_test, label = y_test)
```

```python
predictions = clf.predict(m_X_test)
```

```python
predictions
```

```python
accuracy_score (y_test, predictions)
```

**Categorise columns**

```python
df = pd.read_csv("students_adaptability_level_online_education.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
def category_columns(df):
    columns = df.columns
    for c in columns:
        if c == "Adaptivity Level":
            le = LabelEncoder()
            le.fit(df[c])
            df[c]= le.transform(df[c])
            continue
        df [c] = df[c].astype("category")
    return df
```

```python
category_df = category_columns(df)
```

```python
category_df.head()
```

```python
category_df.dtypes
```

```python
X = category_df.drop (columns="Adaptivity Level")
y = category_df["Adaptivity Level"]
```

```python
X_train,X_test,y_train,y_test = train_test_split (X, y, test_size=0.3, random_state=42)
```

```python
m_X_Train = xgb.DMatrix(data = X_train, label = y_train, enable_categorical=True)
```

```python
clf = xgb.train(
params = {
     "objective": "multi:softmax",
    "num_class": 3,
    "max_depth": 10
    },
    dtrain = m_X_Train
)
```
